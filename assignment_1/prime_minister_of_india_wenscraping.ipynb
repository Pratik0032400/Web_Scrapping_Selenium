```python
driver.get("https://www.jagranjosh.com/")
```

```python
gk=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div/div[1]/div/div[6]/div/div[1]/header/div[3]/ul/li[9]/a')
gk.click()
```

```python
ps=driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a')
ps.click()
```

```python
name = []
names=driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[2]/p')
for i in names:
    name.append(i.text)
```

```python
df=pd.DataFrame({'NAME':name})
df
```

```python
dob = []
dobs=driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[3]/p')
for i in dobs:
    dob.append(i.text)
                          
```

```python
df=pd.DataFrame({'BORN_DEATH':dob})
df
```
