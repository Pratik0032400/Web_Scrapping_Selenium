```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
web='https://chromedriver.chromium.org/downloads'
```

```python
driver=webdriver.Chrome(r"web")
```

```python
driver.get("https://www.motor1.com/")
```

```python
s_1=driver.find_element(By.XPATH,'//a[@href="/features/"]')
s_1.click()
```

```python
s_2=driver.find_element(By.XPATH,'//a[@href="/features/category/lists/"]')
s_2.click()
```

```python
s_3=driver.find_element(By.XPATH,'//a[@href="/features/308149/most-expensive-new-cars-ever/"]')
s_3.click()
```

```python
name = []
```

```python
name_tags=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in name_tags:
    nt=(i.text)
    name.append(nt)
```

```python
len(name)
```

```python
prices = []
```

```python
price_tag=driver.find_elements(By.XPATH,"//strong")
for i in price_tag:
    prices.append(i.text)
```

```python
len(prices)
```

```python
df=pd.DataFrame({'CAR':name,'PRICE':prices})
df
```
