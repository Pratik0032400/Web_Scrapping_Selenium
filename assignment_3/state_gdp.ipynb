```python
gdp_o=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
gdp_o.click()
```

```python
rank = []
state = []
gsdp1819 = []
gsdp1920 = []
share = []
gdpbillion = []
```

```python
ra=driver.find_elements(By.XPATH,'//td[@class="data1"]')
for i in ra[0:32]:
    rank.append(i.text)
```

```python
st=driver.find_elements(By.XPATH,'//td[@class="name"]')
for i in st[0:32]:
    state.append(i.text)
```

```python
gsdp_1819=driver.find_elements(By.XPATH,'//td[@class="data sorting_1"]')
for i in gsdp_1819[0:32]:
    gsdp1819.append(i.text)
```

```python
gsdp_1920=driver.find_elements(By.XPATH,'//tr[@class="odd"]/td[3]')
for i in gsdp_1920[0:32]:
    gsdp1920.append(i.text)
```

```python
len(gsdp1920)
```

```python
sh=driver.find_elements(By.XPATH,'//tr[@class="odd"]/td[5]')
for i in sh[0:32]:
    share.append(i.text)
```

```python
g_billion=driver.find_elements(By.XPATH,'//tr[@class="odd"]/td[6]')
for i in g_billion[0:32]:
    gdpbillion.append(i.text)
```
