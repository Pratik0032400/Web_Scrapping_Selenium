```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
import time
import requests
```

```python
web='https://chromedriver.chromium.org/downloads'
```

```python
driver=webdriver.Chrome(r'web')
```

```python
driver.get("https://github.com/")
```

```python
se=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
se.click()
```

```python
seen=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a')
seen.click()
```

```python
rep_tit = []
rep_des = []
con_con = []
lan_use = []
```

```python
rept=driver.find_elements(By.XPATH,'//span[@class="text-normal"]')
for i in rept[0:10]:
    rep_tit.append(i.text)
```

```python
df=pd.DataFrame({'a':rep_tit})
df
```

```python
repd=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in repd[0:10]:
    rep_des.append(i.text)
```

```python
df=pd.DataFrame({'c':rep_des})
df
```

```python
conn=driver.find_elements(By.XPATH,'//a[@class="Link--muted d-inline-block mr-3"]')
for i in conn[0:10]:
    con_con.append(i.text)
```

```python
df=pd.DataFrame({'d':con_con})
df
```

```python
lanu=driver.find_elements(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]')
for i in lanu[0:10]:
    lan_use.append(i.text)
```

```python
df=pd.DataFrame({'a':lan_use})
df
```

```python
df=pd.DataFrame({'REPOSITORY TITLE':rep_tit,'REPOSITORY DESCRIPTION':rep_des,'CONTRIBUTE COUNT':con_con,'LANGUAGE USE':lan_use})
df
```
