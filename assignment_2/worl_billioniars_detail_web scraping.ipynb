```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
web='https://chromedriver.chromium.org/downloads'
```

```python
driver=webdriver.Chrome(r"web")
```

```python
driver.get("https://www.forbes.com")
time.sleep(3)
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[1]/header/nav/div[1]/div/div/div')
search.click()
```

```python
sw=driver.find_element(By.XPATH,'/html/body/div[1]/header/nav/div[1]/div/div/div[2]/ul/li[1]/div[1]/a')
sw.click()
```

```python
world_billionaires = driver.find_element(By.XPATH,'/html/body/div[1]/main/div/section/section[1]/div/div/div[1]/div/div[1]/div[1]/div[2]/a/h2')
world_billionaires.click()
```

```python
rank = []
rank_tag=driver.find_elements(By.XPATH,'//div[@class="rank"]')
for i in rank_tag[0:20]:
    rt=(i.text)
    rank.append(rt)
```

```python
name = []
name_tag=driver.find_elements(By.XPATH,'//div[@class="personName"]')
for i in name_tag[0:20]:
    na=(i.text)
    name.append(na)
```

```python
net = []
net_tag=driver.find_elements(By.XPATH,'//div[@class="netWorth"]')
for i in net_tag[0:20]:
    nets=(i.text)
    net.append(nets)    
```

```python
age = []
country = []
source = []
industries = []
```

```python
age_tag=driver.find_elements(By.XPATH,'//div[@class="age"]')
for i in age_tag[0:20]:
    at=(i.text)
    age.append(at)
```

```python
country_tag=driver.find_elements(By.XPATH,'//div[@class="countryOfCitizenship"]')
for i in country_tag[0:20]:
    ct=(i.text)
    country.append(ct)
source_tag=driver.find_elements(By.XPATH,'//span[@class="source-text"]')
for i in source_tag[0:20]:
    st=(i.text)
    source.append(st)
ind_tag=driver.find_elements(By.XPATH,'//div[@class="category"]')
for i in ind_tag[0:20]:
    it=(i.text)
    industries.append(it)
```

```python
df=pd.DataFrame({'RANKING':rank,'NAME':name,'AGE':age,'COUNTRY':country,'SOURCE':source,'INDUSTRY':industries})
df
```
