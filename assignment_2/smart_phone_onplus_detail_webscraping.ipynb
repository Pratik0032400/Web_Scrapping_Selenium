Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
import time
import requests
```

```python
web='https://chromedriver.chromium.org/downloads'
```

```python
driver=webdriver.Chrome(r'web')
```

```python
driver.get("https://www.flipkart.com/")
```

```python
ca=driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button')  #login cancel
ca.click()
```

```python
s_input=driver.find_element(By.XPATH,'//input[@class="_3704LK"]')
s_input.send_keys('smart phone')
```

```python
search=driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]')
search.click()
time.sleep(2)
```

```python
product_url = []
url=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in url:
    product_url.append(i.get_attribute("href"))
    time.sleep(3)
```

```python
Brand=[]
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand=driver.find_element(By.XPATH,'//div[@class="_3nWYNs"]')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
```

```python
Model_name=[]
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand=driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        Model_name.append(brand.text)
    except NoSuchElementException:
        Model_name.append('-')
```

```python
colour=[]   #colour
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        c_tag=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        colour.append(c_tag.text)
    except NoSuchElementException:
        colour.append('-')
```

```python
p_camera = []  #primary camera
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        p_tag=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div/div[2]/div[9]/div[5]/div/div[2]/div/div[5]/table/tbody/tr[2]/td[2]/ul/li')
        p_camera.append(p_tag.text)
    except NoSuchElementException:
        p_camera.append('-')
```

```python
s_c = []   #secondary camera
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        s_tag=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[5]/table/tbody/tr[5]/td[2]/ul/li')
        s_c.append(s_tag.text)
    except NoSuchElementException:
        s_c.append('-')
```

```python
ds = []    #dispaly size
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        ds_tag=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[2]/table/tbody/tr[1]/td[2]/ul/li')
        ds.append(ds_tag.text)
    except NoSuchElementException:
        ds.append('-')
```

```python
bc = []   #battery capacity
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        bc_tag=driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[10]/table/tbody/tr/td[2]/ul/li')
        bc.append(bc_tag.text)
    except NoSuchElementException:
        bc.append('-')
```

```python
prs = []   #price
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        prs_tag=driver.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        prs.append(prs_tag.text)
    except NoSuchElementException:
        prs.append('-')
```

```python
df=pd.DataFrame({'BRAND':Brand,'MODEL NAME':Model_name,'COLOUR':colour,'PRIMARY CAMERA':p_camera,'SECONDARY CAMERA':s_c,'DISPLAY':ds,'BATTERY':bc,'PRICE':prs,'PRODUCE_URL':product_url})
df
```
